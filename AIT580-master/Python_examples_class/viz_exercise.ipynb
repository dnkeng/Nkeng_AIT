```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/EmployeeAttrition.csv")
```

```python
df.head()
```

```python
stat = df[['Age', "DailyRate", "DistanceFromHome", "Education", 
            "WorkLifeBalance", "RelationshipSatisfaction", "TotalWorkingYears"]] # way to cut a column as a data frame
```

```python
stat.head()
```

```python
corr = stat.corr()
```

```python
corr
```

```python
cmap = sns.diverging_palette(220, 10, as_cmap=True)
plt.figure(figsize=(10, 10))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, cmap=cmap, vmax=.4, center=0,
            square=True, linewidths=.1)
```
